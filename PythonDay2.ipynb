```python
# pip install selenium

from selenium import webdriver # 브라우저 컨트롤
from bs4 import BeautifulSoup as bs # 브라우저 정제
import pandas as pd
```

```python
driver = webdriver.Chrome('chromedriver.exe')
url = "https://www.melon.com/chart/"
driver.get(url)

html = driver.page_source
```

```python
html
```

```python
soup = bs(html)
```

```python
songs = soup.select('div.wrap_song_info')
songs
```

```python
title = soup.select('div.wrap_song_info > div.rank01 > span > a')[0].text
title
```

```python
singer = soup.select('div.wrap_song_info > div.rank02 > a')[0].text
singer
```

```python
song_list = []
rank = 1

for i in range(100):
    singer = soup.select('div.wrap_song_info > div.rank02 > a')[i].text
    title = soup.select('div.wrap_song_info > div.rank01 > span > a')[i].text
    song_list.append(['멜론', rank, title, singer])
    rank += 1
    
song_list
```

```python
df = pd.DataFrame(song_list, columns = ['사이트', '순위', '제목' , '가수'])
df
```

```python
df.to_excel('melon.xlsx', index = False)
```
