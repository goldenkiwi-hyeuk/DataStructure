```python
# 여러개의 그래프 그리기
data_1 = {'data_x':[1,2,3,4,5], 'data_y': [2,3,5,10,8]}
data_2 = {'data_x':[1,2,3,4,5], 'data_y': [3,5,9,7,7]}

plt.plot('data_x','data_y', data = data_1, label = 'Price ($)')
plt.plot('data_x','data_y', data = data_2, label = 'Demand (#)')
plt.xlabel('X축', fontdict = {'color' : 'b', 'size' : 'x-large'})
plt.ylabel('Y축', fontdict = {'color' : 'r', 'size' : 12})
# plt.legend() # 범례 표시는 1열이 기본
plt.legend(ncol= 2) # ncol 옵션을 활용하여 열이 2개로 표현
plt.show()
```

```python
# 내꺼 버전
# 세개의 선을
# 빨강 실선 [[1,2,3,4], [2,2,5,3,3,4,3]]  리플
# 그린 점선 [[1,2,3,4], [2,3,3,6]]        이더리움
# 블루 삼각형 [[1,2,3,4], [2,3,5,10]]     비트코인

# 범례를 표시하세요
data1 = {'data_x' : [1,2,3,4], 'data_y' :[2,2.5,3.3,4.3]}
data2 = {'data_x' : [1,2,3,4], 'data_y': [2,3,3,6]}
data3 = {'data_x' : [1,2,3,4], 'data_y': [2,3,5,10]}
plt.plot('data_x', 'data_y', 'r-', data = data1,  label = '리플')
plt.plot('data_x', 'data_y', 'g--', data = data2,  label = '이더리움')
plt.plot('data_x', 'data_y', 'b^', data = data3,  label = '비트코인')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.legend()
plt.show()
```

```python
# 강사님 버전
# 세개의 선을
# 빨강 실선   [1,2,3,4],[2, 2.5, 3.3, 4.3]  리플
# 그린 점선   [1,2,3,4],[2, 3,   3,   6]    이더리움
# 블루 삼각형 [1,2,3,4],[2, 3,   5,   10]   비트코인

# 범례를 표시하세요

plt.plot([1,2,3,4],[2, 2.5, 3.3, 4.3], 'r-', label = '리플')
plt.plot([1,2,3,4],[2, 3,   3,   6]  , 'g:', label = '이더리움')
plt.plot([1,2,3,4],[2, 3,   5,   10], 'b^', label = '비트코인')
plt.legend(ncol = 3)
plt.show()
```

```python
import numpy as np # 계산 관련 라이브러리
a = np.arange(0,2,0.2) # 0부터 2까지 0.2씩 리스트 생성

plt.plot(a,a, 'r--')
plt.plot(a,a**2, 'bo', markersize = 20)
plt.plot(a,a**3, 'g-', linewidth = 10)

# plt.grid(True) # x,y축 그리드 모두 설정
plt.grid(True, axis = 'y') # y축만 그리드 설정
plt.show()
```

```python
# 타이틀 설정

import numpy as np # 계산 관련 라이브러리
a = np.arange(0,2,0.2) # 0부터 2까지 0.2씩 리스트 생성

plt.plot(a,a, 'r--')
plt.plot(a,a**2, 'bo', markersize = 20)
plt.plot(a,a**3, 'g-', linewidth = 10)

# plt.grid(True) # x,y축 그리드 모두 설정
plt.grid(True, axis = 'y') # y축만 그리드 설정

# plt.title('그래프 제목')
plt.title('그래프 제목', loc = 'right', fontdict = {'fontsize': 16, 'fontweight': 'bold'}) # loc 생략시 가운데
plt.show()
```

```python
# 막대 그래프

z = np.arange(3)
x = ['2020','2021','2022']
y = [100,400,900]
plt.bar(z,y, width = 0.6, align = 'edge', color = 'green' ) 
plt.xticks(z,x) # x축에서 z데이터를 x데이터로 치환

plt.show()
```

```python
# 막대 그래프

z = np.arange(3)
x = ['2020','2021','2022']
y = [100,400,900]
plt.barh(z,y, color = 'green' ) 
plt.yticks(z,x) # z를 x로 치환

plt.show()
```
